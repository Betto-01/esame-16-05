```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

path = 'https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/life.csv'
```

```python
X = df.drop(columns='Life_exp')
y = df['Life_exp']
```

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# standardized_data = sc.fit_transform(X_)

# X = standardized_data
# X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=667
                                                    )
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_clf = DecisionTreeRegressor(max_depth=4)
dt_clf.fit(X_train, y_train)
dt_clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_clf = RandomForestRegressor(n_estimators=200)
rf_clf.fit(X_train, y_train)
rf_clf.score(X_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression
model_slim = LinearRegression()
model_slim.fit(X_train, y_train)
model_slim.score(X_test, y_test)
```

```python
y_pred = model_slim.predict(X_test)
```

```python
model_slim.intercept_
```

```python
model_slim.coef_
```

```python
import numpy as np
length = y_pred.shape[0] #  
x = np.linspace(0,length,length)


plt.figure(figsize=(30,15))
plt.plot(x, y_test, label='reale')
plt.plot(x, y_pred, label='predetto')
plt.legend(loc=2);
```

```python
confronto = pd.DataFrame(data = [y_pred,y_test])
confronto = confronto.T
confronto.columns = ['predetto', 'reale']
confronto
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('R2_score: ', round(r2score,4))
print('MAE: ', round(mae,2))
print('MSE: ', round(mse,2))
print('RMSE: ', round(rmse,2))
```
